```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {"Country": ["China", "India", "USA", "Indonesia", "Pakistan"],
    "Population (millions)": [1411, 1380, 331, 273, 229]}

df = pd.DataFrame(data)
df
```

```python
plt.bar(df["Country"], df["Population (millions)"])
plt.title("World Population by Country")
plt.xlabel("Country")
plt.ylabel("Population (millions)")
plt.show()
```
